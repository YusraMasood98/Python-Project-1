```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pl
data=pd.read_csv("House price.csv")
area=np.array(data.iloc[:,0])
price=np.array(data.iloc[:,1])
area2=np.array(area)*0.0001
price2=np.array(price)*289.37
x=area2
y=price2
pl.title("Matplotlib demo")
pl.xlabel("Area in meter square")
pl.ylabel("Price in pkr")
pl.plot(x,y)
pl.show()
```
